````python
import pandas as pd

import matplotlib.pyplot as plt

from wordcloud import WordCloud
````

````python
gis_word = r'https://raw.githubusercontent.com/oderofrancis/rona/main/gis_word_cloud.csv'
````

````python
gis_word = pd.read_csv(gis_word)

gis_word.columns = ['words']

gis_word.head()
````

````python
gis_word = " ".join(cat.split()[0] for cat in gis_word.words)
````

````python
wordcloud = WordCloud(max_words=100).generate(gis_word)
````

````python
plt.imshow(wordcloud, interpolation="quadric")

plt.axis("off")

plt.show()
````

The available interpolation methods are: 

```'none', 'nearest', 'bilinear', 'bicubic', 'spline16', 'spline36', 'hanning', 'hamming', 'hermite', 'kaiser', 'quadric', 'catrom', 'gaussian', 'bessel', 'mitchell', 'sinc', 'lanczos'.```
